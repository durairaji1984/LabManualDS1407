### read the xlsx data using pd.read_excel function ###

```python
data = pd.read_excel(path_to_data)
```

```python
data.head()
```

```python
data.head(10)
```

### Dropping columns in the data


```python
df_dropped = data.drop('S.No.', axis=1)
df_dropped.head()
```

### Drop multiple columns using the following code

```python
df_dropped_multiple = data.drop(['S.No.', 'Tamils and Technology'], axis=1)
df_dropped_multiple.head()
```

### Dropping rows in the data

```python
df_row_dropped = data.drop(2, axis=0)
df_row_dropped.head()
```

### Renaming a column in the dataset data.columns

```python
df_renamed = data.rename(columns={'Reg.No.': 'Register Number'})
df_renamed.head() 
```

### Select columns with specific data types

```python
integer_data = data.select_dtypes('int')
integer_data.head()
```

```python
integer_data = data.select_dtypes('object')
integer_data.head()
```
