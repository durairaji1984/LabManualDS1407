## Histogram

### Create a histogram of sepal length,sepal.width,petal.length and petal.width

```python
plt.subplot(2, 2, 1)
plt.hist(X['sepal.length'], bins=10)
plt.xlabel('Sepal length')
plt.ylabel('Frequency')
plt.title('Histogram of Sepal length')
plt.subplot(2, 2, 2)
plt.hist(X['sepal.width'], bins=10)
plt.xlabel('Sepal Width')
plt.ylabel('Frequency')
plt.title('Histogram of Sepal Width')
```

```python

plt.subplot(2, 2, 3)
plt.hist(X['petal.length'], bins=10)
plt.xlabel('Petal length')
plt.ylabel('Frequency')
plt.title('Histogram of Petal length')
plt.subplot(2, 2, 4)
plt.hist(X['petal.width'], bins=10)
plt.xlabel('Petal width')
plt.ylabel('Frequency')
plt.title('Histogram of Petal width')
plt.show()
```

## scatter plot

### Create a scatter plot of sepal length vs sepal width

```python
#checking for Null values
iris.isnull().sum()
```

```python
encoder = LabelEncoder()
iris['variety'] = encoder.fit_transform(iris['variety'])
```

```python
#finding different class labels 
np.unique(iris['variety'])
```

```python
x1=X.to_numpy()
x1
```

```python
y1=iris['variety'].to_numpy()
y1
```

```python
classes = ['Setosa', 'Versicolor', 'Virginica']
scatter = plt.scatter(X['sepal.length'], X['sepal.width'], c=iris.variety)
plt.legend(handles=scatter.legend_elements()[0], labels=classes)
plt.xlabel('Sepal length')
plt.ylabel('Sepal width')
plt.title('Scatter plot of Sepal length vs Sepal width')
plt.show()
```

```python
scatter = plt.scatter(X['petal.length'], X['petal.width'], c=iris.variety)
plt.legend(handles=scatter.legend_elements()[0], labels=classes)
plt.xlabel('Petal length')
plt.ylabel('Petal width')
plt.title('Scatter plot of Petal length vs Petal width')
plt.show()
```

## line plot

### Create a line plot of sepal length,sepal.width,petal.length and petal.width over time

```python
plt.subplot(2, 2, 1)
plt.plot(X['sepal.length'])
plt.xlabel('Index')
plt.ylabel('Sepal length')
plt.title('Line Plot of Sepal length')
plt.subplot(2, 2, 2)
plt.plot(X['sepal.width'])
plt.xlabel('Index')
plt.ylabel('Sepal Width')
plt.title('Line Plot of Sepal Width')
```

```python
plt.subplot(2, 2, 3)
plt.plot(X['petal.length'])
plt.xlabel('Index')
plt.ylabel('Petal length')
plt.title('Line Plot of Petal length')
plt.subplot(2, 2, 4)
plt.plot(X['petal.width'])
plt.xlabel('Index')
plt.ylabel('Petal width')
plt.title('Line Plot of Petal width')
plt.show()
```

## Bar Plot

### Create a bar plot of the average sepal length for each target class


```python
avg_sepal_length = [x1[y1 == i, 0].mean() for i in range(3)]
target_names = classes
plt.bar(target_names, avg_sepal_length)
plt.xlabel('Target class')
plt.ylabel('Average sepal length')
plt.title('Bar plot of Average Sepal length for each Target class')
plt.show()
```

## Pie Chart

### Create a pie chart of the proportion of each target class

```python
proportions = [sum(y1 == i) for i in range(3)]
```

```python
plt.pie(proportions, labels=classes)
plt.title('Pie chart of Target class proportions')
plt.show()
```
